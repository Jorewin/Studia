```python
# requires-python = ">3.8.0"

# python requirements

"""
numpy==1.24.1
sympy==1.11.1
"""

import random
import numpy
from sympy import randprime
from IPython.display import Markdown
from ipywidgets import Button, Checkbox, IntText, Output
```

```python
class KeyGenerator:
    def __init__(
        self,
        prime_widget: IntText,
        base_widget: IntText,
        generate_parameters_widget: Checkbox,
        start_widget: Button,
        output_widget: Output
    ):
        self.prime_widget = prime_widget
        self.base_widget = base_widget
        self.generate_parameters_widget = generate_parameters_widget
        self.start_widget = start_widget
        self.output_widget = output_widget
        self.widgets = [
            prime_widget,
            base_widget,
            generate_parameters_widget,
            start_widget
        ]
        
        self.generate_keys = output_widget.capture()(self.generate_keys)
        
    def disable_widgets(self):
        for widget in self.widgets:
            widget.enabled = False
            
    def enable_widgets(self):
        for widget in self.widgets:
            widget.enabled = True
        
    def generate_keys(self):
        self.output_widget.clear_output()
        self.disable_widgets()
        self.start_widget.description = "Computing"
        
        prime = self.prime_widget.value
        base = self.base_widget.value
        
        if self.generate_parameters_widget.value:
            prime = randprime(100, 1000)
            self.prime_widget.value = prime
            required_set = set(range(1, prime))
            
            for i in range(prime - 1, 0, -1):
                actual_set = set(i**j%prime for j in range(prime - 1))
                
                if actual_set == required_set:
                    base = i
                    self.base_widget.value = base
                    break
                    
        alice_secret = random.randint(100, 1000)
        bob_secret = random.randint(100, 1000)
        print("Alice's secret: ", alice_secret)
        print("Bob's secret:   ", bob_secret)
        
        x = base**alice_secret%prime
        y = base**bob_secret%prime
        print("Alice sends:    ", x)
        print("Bob sends:      ", y)
        
        alice_key = y**alice_secret%prime
        bob_key = x**bob_secret%prime
        print("Alice got:      ", alice_key)
        print("Bob got:        ", bob_key)
        
        self.start_widget.description = "Start"
        self.enable_widgets()
```

```python
prime_widget = IntText(description="Prime:")
base_widget = IntText(description="Base:")
generate_parameters_widget = Checkbox(value=True, description="Generate parameters:")
start_widget = Button(description="Start")
output_widget = Output()

key_generator = KeyGenerator(
    prime_widget,
    base_widget,
    generate_parameters_widget,
    start_widget,
    output_widget
)

start_widget.on_click(lambda _: key_generator.generate_keys())

display(
    prime_widget,
    base_widget,
    generate_parameters_widget,
    start_widget,
    output_widget
)
```
